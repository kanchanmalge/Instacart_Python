# Create a spending flag for each user based on the average price across all their orders using the following criteria:

1. If the mean of the prices of products purchased by a user is lower than 10, then flag them as a Low spender.
2. If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a High spender.



```python
# Creating average_price column using transform()

df_orders_products_merged['average_price'] = df_orders_products_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_orders_products_merged.head()
```

```python
# Rounding averages 

df_orders_products_merged['average_price'] = df_orders_products_merged['average_price'].round(2)
```

```python
df_orders_products_merged.head()
```

```python
# Creating a 'spending_flag' flag

df_orders_products_merged.loc[df_orders_products_merged['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Creating a 'spending_flag' flag

df_orders_products_merged.loc[df_orders_products_merged['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_orders_products_merged.head()
```

```python
# Check the frequency

df_orders_products_merged['spending_flag'].value_counts(dropna = False)
```

# Step 7
Create an order frequency flag that marks the regularity of a users ordering behavior according to the median in the days_since_prior_order column. The criteria for the flag should be as follows:

1. If the median of days_since_prior_order is higher than 20, then the customer should be labeled a Non-frequent customer.
2. If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a Regular customer.
3. If the median is lower than or equal to 10, then the customer should be labeled a Frequent customer.



```python
# Creating median_prior_orders column using transform()

df_orders_products_merged['median_prior_orders'] = df_orders_products_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_orders_products_merged.head()
```

```python
# Creating a 'order_frequency_flag' flag

df_orders_products_merged.loc[df_orders_products_merged['median_prior_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
# Creating a 'order_frequency_flag' flag

df_orders_products_merged.loc[(df_orders_products_merged['median_prior_orders'] > 10) & (df_orders_products_merged['median_prior_orders'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
# Creating a 'order_frequency_flag' flag

df_orders_products_merged.loc[df_orders_products_merged['median_prior_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
df_orders_products_merged.head()
```

```python
# Check the frequency

df_orders_products_merged['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Check the null values

print('Records in df_orders_products_merged with null values in column order_frequency_flag:')
df_orders_products_merged[df_orders_products_merged['order_frequency_flag'].isnull()][['user_id', 'days_since_prior_order', 'median_prior_orders', 'order_frequency_flag']]
```
